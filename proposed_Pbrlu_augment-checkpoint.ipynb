```python
#Multi-Scale Patch Blur Augmentation for Grayscale NIR Images

#Explanation
#Input Parameters:

#image: The input NIR grayscale image.
#bboxes: A list of bounding boxes around pedestrians in the format (xmin, ymin, xmax, ymax).
#scales: A list of scales to determine the size of the blur patches relative to the bounding box dimensions.
#edge_extension: The number of pixels to extend beyond the bounding box edges.
#draw_boxes: A boolean flag to indicate whether to draw bounding boxes around the blurred areas.
#Process:

#Convert the grayscale image to a 3-channel RGB format (to maintain consistency with HSV processing).
#For each scale in the scales list, and for each bounding box in the bboxes list:
#Calculate the width and height of the blur patch based on the scale and bounding box dimensions.
#Extend the region slightly beyond the bounding box using edge_extension pixels.
#Randomly select a position for the blur patch within the extended region.
#Apply Gaussian blur to the selected patch.
#Optionally draw a bounding box around the blurred area if draw_boxes is True.
#Convert the image back to grayscale format after applying all blurs.
#Benefits
#Increased Coverage: Including the edge of the bounding box and a small area around it ensures the augmentation affects not just the object but also its immediate context.
#Contextual Relevance: The method still focuses on regions relevant to the object of interest, providing more meaningful augmentation.
#Enhanced Variability: The random positioning within the extended region introduces additional variability, helping the model generalize better.
#This approach leverages the ground truth bounding boxes and slightly extends the augmentation region to cover the edges and a small area around them, thereby improving the model's robustness and accuracy in detecting pedestrians in NIR images.

#one more thing that this novel augmentation method can be good to distinguish the pedestrian from the background.






import cv2
import numpy as np
import random
import matplotlib.pyplot as plt

def multi_scale_patch_blur_augmentation(image, bboxes, scales=[0.1, 0.2, 0.3], edge_extension=10):
    # Convert the image to HSV (though it's grayscale, we still convert to a 3-channel format for consistency)
    hsv_image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    height, width = hsv_image.shape[:2]
    
    # List to store the positions of the blurred patches
    blur_positions = []

    for scale in scales:
        for bbox in bboxes:
            xmin, ymin, xmax, ymax = bbox
            bbox_width = xmax - xmin
            bbox_height = ymax - ymin
            
            blur_width = int(scale * bbox_width)
            blur_height = int(scale * bbox_height)
            
            # Extend the region slightly beyond the bounding box
            extended_xmin = max(0, xmin - edge_extension)
            extended_ymin = max(0, ymin - edge_extension)
            extended_xmax = min(width, xmax + edge_extension)
            extended_ymax = min(height, ymax + edge_extension)
            
            # Randomly select a position for the blur patch within the extended region
            x = random.randint(extended_xmin, extended_xmax - blur_width)
            y = random.randint(extended_ymin, extended_ymax - blur_height)
            
            blur_box = hsv_image[y:y+blur_height, x:x+blur_width]
            blurred_box = cv2.GaussianBlur(blur_box, (15, 15), 0)
            hsv_image[y:y+blur_height, x:x+blur_width] = blurred_box
            
            # Store the position of the blur
            blur_positions.append((x, y, blur_width, blur_height))
    
    output_image = cv2.cvtColor(hsv_image, cv2.COLOR_RGB2GRAY)
    return output_image, blur_positions

# Example usage
image = cv2.imread('C:/Users/Nassan/Desktop/Nir/frame_image30.jpg', cv2.IMREAD_GRAYSCALE)
bboxes = [(0, 161, 53, 281), (112, 89, 266, 320)]  # Example bounding boxes
# Apply multi-scale patch blur augmentation
augmented_image, blur_positions = multi_scale_patch_blur_augmentation(image.copy(), bboxes)

# Draw rectangles on a separate copy of the blurred image to show where the blur was applied
blur_with_rectangles = cv2.cvtColor(augmented_image.copy(), cv2.COLOR_GRAY2RGB)
for (x, y, blur_width, blur_height) in blur_positions:
    cv2.rectangle(blur_with_rectangles, (x, y), (x + blur_width, y + blur_height), (0, 255, 0), 2)

# Display the results
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Patch Blur Augmented Image')
plt.imshow(augmented_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Blurred Image with Blur Areas Highlighted')
plt.imshow(blur_with_rectangles)
plt.axis('off')

# Save the resulting images using plt
plt.savefig('C:/Users/Nassan/Desktop/patch_blur_augmentation/NIR_results/NIR-multi_patch_blur/Mul_Patch_blur17.jpg')

plt.show()
```

```python
#To incorporate evaluation metrics into the provided code,
#let's focus on calculating the Structural Similarity Index (SSIM) between the original image and the augmented image after applying
#the multi-scale patch blur augmentation. SSIM is a widely used metric for measuring image quality, which considers luminance, contrast,
#and structure similarity between two images.

import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim

def multi_scale_patch_blur_augmentation(image, bboxes, scales=[0.1, 0.2, 0.3], edge_extension=10):
    hsv_image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    height, width = hsv_image.shape[:2]
    
    blur_positions = []

    for scale in scales:
        for bbox in bboxes:
            xmin, ymin, xmax, ymax = bbox
            bbox_width = xmax - xmin
            bbox_height = ymax - ymin
            
            blur_width = int(scale * bbox_width)
            blur_height = int(scale * bbox_height)
            
            extended_xmin = max(0, xmin - edge_extension)
            extended_ymin = max(0, ymin - edge_extension)
            extended_xmax = min(width, xmax + edge_extension)
            extended_ymax = min(height, ymax + edge_extension)
            
            x = random.randint(extended_xmin, extended_xmax - blur_width)
            y = random.randint(extended_ymin, extended_ymax - blur_height)
            
            blur_box = hsv_image[y:y+blur_height, x:x+blur_width]
            blurred_box = cv2.GaussianBlur(blur_box, (15, 15), 0)
            hsv_image[y:y+blur_height, x:x+blur_width] = blurred_box
            
            blur_positions.append((x, y, blur_width, blur_height))
    
    output_image = cv2.cvtColor(hsv_image, cv2.COLOR_RGB2GRAY)
    return output_image, blur_positions

def calculate_ssim(image1, image2):
    return ssim(image1, image2, data_range=image2.max() - image2.min())

# Example usage
image = cv2.imread('C:/Users/Nassan/Desktop/Nir/frame_image30.jpg', cv2.IMREAD_GRAYSCALE)
bboxes = [(0, 161, 53, 281), (112, 89, 266, 320)]  # Example bounding boxes

augmented_image, blur_positions = multi_scale_patch_blur_augmentation(image.copy(), bboxes)

# Calculate SSIM between original and augmented images
ssim_value = calculate_ssim(image, augmented_image)

# Draw rectangles on a separate copy of the blurred image
blur_with_rectangles = cv2.cvtColor(augmented_image.copy(), cv2.COLOR_GRAY2RGB)
for (x, y, blur_width, blur_height) in blur_positions:
    cv2.rectangle(blur_with_rectangles, (x, y), (x + blur_width, y + blur_height), (0, 255, 0), 2)

# Display the results including SSIM value
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title(f'Patch Blur Augmented Image\nSSIM: {ssim_value:.4f}')
plt.imshow(augmented_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Blurred Image with Blur Areas Highlighted')
plt.imshow(blur_with_rectangles)
plt.axis('off')

plt.show()
```
